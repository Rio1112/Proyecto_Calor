# Proyecto ecuacion de calor en Python
Utilizando el metodo de diferencias finitas para resolver la ecuacion en dos dimensiones espaciales.

```python
import numpy as np

def ftcs_calor(Th, Tc, kappa, L, a, M, N, h, t_fin):
    # Condiciones iniciales
    T = np.full((M+1, N+1), Tc, dtype=float)
    T[24:27, 24:27] = Th

    #Los otros parametros
    t = 0.0
    dy= a / (N - 1)  #Particiones del ancho
    dx = L / (M - 1)  #Partciones del largo

    #Para la animacion
    its = 0
    tsteps = int(t_fin / h)
    shape = np.zeros((tsteps + 2, M+1, N+1), dtype=float)

    #Aqui se implementa el FTCS
    while t < t_fin:
        #arreglo para hacer las operaciones con FTCS
        Tk = T.copy()
        for i in range(0, M):
            for j in range(0, N):
               if i < 24 or i > 26 or j < 24 or j > 26:
                    Tk[i, j] = (((kappa*h)/(dy*dx)) * ((dy/dx) * (T[i-1, j] + T[i+1, j] - 2*T[i, j])) + ((kappa*h)/(dy*dx)) * ((dx/dy) * (T[i, j+1] + T[i, j-1] - 2*T[i, j])) + T[i, j])

        T = Tk.copy()
        t += h
        its += 1
        shape[its] = T.copy()
    return T, its, shape

def initial_condition(x, y):
    if 0.4 <= x <= 0.6 and 0.4 <= y <= 0.6:
        return 100
    else:
        return 20
```

Para la primera condicion incial:

```python
valores, iteraciones, shape = ftcs_calor(300.0, 20.0, 98.8e-6, 0.3, 0.3, 50, 50, 1e-2, 60.0)
print(iteraciones)
```

Para la segunda condicion inicial:

```python
valores_2, iteraciones_2, shape_2 = ftcs_calor(200.0, 20.0, 98.8e-6, 0.3, 0.3, 50, 50, 1e-2, 60.0)
print(iteraciones_2)
```

Grafico de calentamiento de la placa, sin animar:

```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm

# Grafica 1
plt.figure()
plt.imshow(valores, extent=[0, 30, 0, 30], origin='lower', cmap='hot', interpolation='nearest', vmin=20, vmax=300)
plt.colorbar()
plt.title('Primera condicion')
plt.xlabel('Distancia en x(cm)')
plt.ylabel('Distancia en y(cm)')

# Grafica 2
plt.figure()
plt.imshow(valores_2, extent=[0, 30, 0, 30], origin='lower', cmap='hot', interpolation='nearest', vmin=20, vmax=300)
plt.colorbar()
plt.title('Segunda condicion')
plt.xlabel('Distancia en x(cm)')
plt.ylabel('Distancia en y(cm)')

plt.show()
```

### Animacion 1
Este dura 60 segundos:

```python
from matplotlib.animation import FuncAnimation
from matplotlib.pyplot import imshow
from IPython import display

fig, ax = plt.subplots()
heatmap = ax.imshow(valores, extent=[0, 30, 0, 30], cmap='hot', interpolation='nearest', vmin=20, vmax=300)
cbar = fig.colorbar(heatmap, ax=ax)
plt.xlabel('Distancia en x(cm)')
plt.ylabel('Distancia en y(cm)')

def animate(frame):
    dato = shape[frame, :, :]
    heatmap.set_data(dato)
    return heatmap
    
anim = FuncAnimation(fig, animate, frames = iteraciones, interval = 3)
video = anim.to_html5_video()
html = display.HTML(video)
display.display(html)
plt.close()
```

## Animacion 2
Para comparar, dura misma cantidad de tiempo, pero se cambian las condiciones iniciales:

```python
fig_2, ax_2 = plt.subplots()
heatmap_2 = ax_2.imshow(valores_2, extent=[0, 30, 0, 30], cmap='hot', interpolation='nearest', vmin=20, vmax=300)
cbar_2 = fig_2.colorbar(heatmap_2, ax=ax_2)
plt.xlabel('Distancia en x(cm)')
plt.ylabel('Distancia en y(cm)')

def animate_2(frame):
    dato = shape_2[frame, :, :]
    heatmap_2.set_data(dato)
    return heatmap_2
    
anim = FuncAnimation(fig_2, animate_2, frames = iteraciones_2, interval = 3)
video = anim.to_html5_video()
html = display.HTML(video)
display.display(html)
plt.close()
```
